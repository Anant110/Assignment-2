```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
np.round(X_train.describe(),1)
```

# Effect Of Scaling

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(X_train['Age'],X_train['Billing Amount'])
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scaled['Age'],X_train_scaled['Billing Amount'],color='red')
ax2.set_title('After Scaling')
plt.show()
```

# Logistic Regression

```python
df=pd.read_csv("healthcare_dataset.csv")
```

```python
col1=df[['Age','Billing Amount','Room Number','Test Results']]
```

```python
col1
```

```python
col1['Test Results'].unique()
```

```python
col1.isnull().sum()
```

```python
col1=col1[col1['Test Results']!='Abnormal']
```

```python
col1.head()
```

```python
col1_copy = col1.copy()
col1_copy['Test Results'] = col1_copy['Test Results'].map({'Inconclusive': 0, 'Normal': 1})
```

```python
col1_copy.head()
```

```python
##Split dataset into dependent and independent features
X=col1_copy.iloc[:,:-1]
y=col1_copy.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train1,X_test1,y_train1,y_test1=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
parameter = {
    'penalty': ['l2', 'none'],  # Adjusted penalties for lbfgs solver
    'C': [1, 2, 3, 4, 5, 6, 10, 20, 30, 40, 50],
    'max_iter': [100, 200, 300]
}


```

# Hyperparameter Optimization

```python
from sklearn.model_selection import GridSearchCV

classifier_regression=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_regression.fit(X_train1,y_train1)
```

```python
print(classifier_regression.best_params_)
```

```python
print(classifier_regression.best_score_)
```

```python
##prediction
y_pred=classifier_regression.predict(X_test1)
```

```python
## accuracy Score
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test1)
print(score)
```

```python
print(classification_report(y_pred,y_test1))
```

# Random Forest

```python
df=pd.read_csv("healthcare_dataset.csv")
```

```python
df.info()
```

```python
feature_cols = ['Age','Admission Type','Billing Amount']
```

```python
X = df[feature_cols]
y = df['Gender']
```

```python
X.head()
```

```python
y.head()
```

```python
X['Admission Type'] = X['Admission Type'].map({'Elective': 0, 'Emergency': 0, 'Urgent': 1})
y = y.map({'Male': 0, 'Female': 1})  # Assuming 'Female' is the correct label and fixing the space in ' Female'

y = y.fillna(-1).astype('int64')  # Fill missing values and convert to int64
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
for i in X_train, X_test, y_train, y_test:
    print(len(i))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```
